### Linear Regression:
Linear regression is a statistical method used to model the relationship between a dependent variable (target) and one or more independent variables (features). In the case of Boston housing price prediction, the dependent variable would be the housing price, and the independent variables would be various features such as crime rate, number of rooms, proximity to employment centers, etc.
The basic form of the linear regression equation for predicting the target variable $\hat{Y}$ from the features $X_1, X_2, ..., X_n$ is:

$$\hat{Y} = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + ... + \beta_n X_n + \epsilon$$

**Where:**

* $\hat{Y}$ (pronounced "Y hat") is the predicted value (housing price in this case).
* $X_1, X_2, ..., X_n$ are the independent variables (features).
* $\beta_0, \beta_1, ..., \beta_n$ are the coefficients, representing the weights assigned to each feature.
* $\epsilon$ (pronounced "epsilon") is the error term, which accounts for the difference between the predicted value and the actual value.

### Deep Neural Networks (DNNs):
A deep neural network is a type of artificial neural network (ANN) with multiple layers between the input and output layers. Each layer consists of neurons (or nodes) that perform simple operations on the input data. The deep part of the name comes from having many hidden layers, which allows the network to learn complex patterns in the data.

### Using DNNs for Linear Regression:
To perform linear regression using a deep neural network, you can create a neural network with a single output neuron and no activation function. The output of this neuron would be the predicted value, and the network would learn the weights (coefficients) that minimize the difference between the predicted values and the actual target values.

### Boston Housing Price Prediction:
In the context of Boston housing price prediction, you would construct a deep neural network with an input layer consisting of the features (e.g., crime rate, number of rooms, etc.), a single output neuron for predicting the housing price, and no activation function. You would then train the neural network using an optimization algorithm (e.g., stochastic gradient descent) to minimize the mean squared error between the predicted prices and the actual prices in the training data.

## Conclusion:
By using a deep neural network for linear regression, you can leverage the power of neural networks to learn complex relationships between features and the target variable, potentially leading to more accurate predictions compared to traditional linear regression methods.

# Import Library

```python
# Data analysis and visualization
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

# Preprocessing and evaluation
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler
```

# Load Data

```python
(X_train , y_train), (X_test , y_test) = tf.keras.datasets.boston_housing.load_data(
                                            path = 'boston_housing_npz',
                                            test_split = 0.2,
                                            seed = 42
                                        )
```

# Exploratory Data Analysis

## Initial Observation

```python
# Checking the data shape and type
(X_train.shape, type(X_train)), (X_test.shape, type(X_test)), (y_train.shape, type(y_train)), (y_test.shape, type(y_test)),
```

```python
# Converting Data to DataFrame 
X_train_df = pd.DataFrame(X_train)
y_train_df = pd.DataFrame(y_train)

# Preview the training data
X_train_df.head(10)
```

```python
# View summary of datasets
X_train_df.info()
print('_'*40)
y_train_df.info()
```

## Preprocessing

```python
# Create column transformer
ct = make_column_transformer(
    (MinMaxScaler(), [0, 1, 2, 4, 5, 6, 7, 8, 9, 10, 11, 12])
)

# Normalization and data type change
X_train = ct.fit_transform(X_train).astype('float32')
X_test = ct.transform(X_test).astype('float32')
y_train = y_train.astype('float32')
y_test = y_test.astype('float32')

# Distribution of X_train feature values after normalization
pd.DataFrame(X_train).describe()
```

## Model, Predict, Evaluation

```python
# Reserve data for validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

### Creating the Model and Optimizing the Learning Rate
learning rate = 0.01, batch_size = 32, dense_layers = 2, hidden_units for Dense_1 layer= 10, hidden_units for Dense_2 layer = 100

```python
# Set random seed
tf.random.set_seed(42)

# Building the model
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=10, activation='relu', input_shape=(X_train.shape[1],), name='Dense_1'),
  tf.keras.layers.Dense(units=100, activation='relu', name='Dense_2'),
  tf.keras.layers.Dense(units=1, name='Prediction')
])

# Compiling the model
model.compile(
    loss = tf.keras.losses.mean_squared_error,
    optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.01),
    metrics = ['mse']
)

# Training the model
history = model.fit(
    X_train,
    y_train,
    batch_size=32,
    epochs=50,
    validation_data=(X_val, y_val)
)
```

## Model Evaluation

```python
# Preview the mean value of training and validation data
y_train.mean(), y_val.mean()
```

```python
# Evaluate the model on the test data
print("Evaluation on Test data \n")
loss, mse = model.evaluate(X_test, y_test, batch_size=32)
print(f"\nModel loss on test set: {loss}")
print(f"Model mean squared error on test set: {(mse):.2f}")
```

```python
# Plot the loss curves
pd.DataFrame(history.history).plot(figsize=(6, 4), xlabel="Epochs", ylabel="Loss", title='Loss Curves')
plt.show()
```

## Model Prediction

```python
# Make predictions
y_pred = model.predict(X_test)

# View the first prediction
y_pred[0]
```
